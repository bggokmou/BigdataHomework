```python
import pandas as pd
import numpy as np

path = 'https://github.com/dongupak/DataML/raw/main/csv/'
file = path + 'vehicle_prod.csv'
df = pd.read_csv(file, index_col=0)

# 모든 숫자형 열의 합계 계산
sums = df.select_dtypes(np.number).sum().rename('Total')

# 모든 숫자형 열의 평균 계산
averages = df.select_dtypes(np.number).mean().rename('Average')

# 합계와 평균을 DataFrame으로 변환
sums_df = pd.DataFrame(sums).transpose()
averages_df = pd.DataFrame(averages).transpose()

# 원본 DataFrame에 합계와 평균 DataFrame 연결
total_df = pd.concat([df, sums_df, averages_df])

# 소수점 둘째 자리에서 반올림
total_df = total_df.round(2)

# 결과 DataFrame 표시
total_df
```
